```python
from os import stat
from CoolProp.CoolProp import PropsSI
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


###### getting the states:

def getstates(P=None,T=None,Q=None,H=None,S=None,fluid='Water'):
  states = {} # Initialize the states dictionary here
  if P is not None and Q is not None:
    states['T'] = PropsSI('T','P',P,'Q',Q,fluid)
    states['S'] = PropsSI('S','P',P,'Q',Q,fluid)
    states['h'] = PropsSI('H','P',P,'Q',Q,fluid)
    states['Q'] = Q
    states['P'] = P
  elif P is not None and S is not None:
    states['T'] = PropsSI('T','P',P,'S',S,fluid)
    states['h'] = PropsSI('H','P',P,'S',S,fluid)
    states['S'] = S
    states['P'] = P
  elif P is not None and H is not None:
    states['T'] = PropsSI('T','P',P,'H',H,fluid)
    states['S'] = PropsSI('S','P',P,'H',H,fluid)
    states['h'] = H
    states['P'] = P
  return states

###### Rankine Calculation
def rankine_cycle(P_high,P_low,eta_turbine,eta_pump,fluid='Water'):

  #State1: Saturated liq at condensor pressure
  State1=getstates(P=P_low,Q=0,fluid=fluid)

  #State 2s: ideal pump outlet
  h2s = PropsSI('H','P',P_high,'S',State1['S'],fluid)
  h2=State1['h']+(h2s-State1['h'])/eta_pump
  State2s = getstates(P=P_high,H=h2s,fluid=fluid)
  # State2: real pump outlet state
  State2=getstates(P=P_high,H=h2,fluid=fluid)

  #State3: sat. vap at boiler
  State3 = getstates(P=P_high,Q=1,fluid=fluid)

  #State4s: ideal turbine outlet
  h4s = PropsSI('H','P',P_low,'S',State3['S'], fluid) # Corrected fluid argument
  h4=State3['h']-(State3['h']-h4s)*eta_turbine
  State4s=getstates(P=P_low,H=h4s,fluid=fluid)
  #State4: real turbine outlet
  State4=getstates(P=P_low,H=h4,fluid=fluid)


  #Energy balance
  qin = State3['h']-State2['h']
  wpump = State2['h']-State1['h']
  wturb = State3['h']-State4['h']
  qout = State4['h']-State1['h']
  wnet = wturb-wpump
  eff = wnet/qin*100

  return {'actual_states': [State1,State2,State3,State4],'ideal_states': [State1,State2s,State3,State4s],'pump work': wpump, 'turbine work': wturb, 'net_work':wnet,'heat added': qin,'Effeciency':eff}

####Using pandas to store data
result = rankine_cycle(P_high=3e6,P_low=10e3,eta_turbine=0.85,eta_pump=0.85)
df_actualstates = pd.DataFrame(result['actual_states'],index=['State1','State2','State3','State4'])
df_idealstates = pd.DataFrame(result['ideal_states'],index=['State1','State2s','State3','State4s'])
print('Actual cycle results')
print(df_actualstates)
print('Ideal cycle results')
print(df_idealstates)
summary = {'pump work':result['pump work'],'turbine work':result['turbine work'],'net work':result['net_work'],'heat added':result['heat added'],'Effeciency':result['Effeciency']}
df_summary = pd.DataFrame(summary,index=[''])
print(df_summary)


####generate saturation dome
plt.figure(figsize=(10, 6))
Prange = np.logspace(np.log10(1e3), np.log10(22e6), 200)
Tsat1 = [PropsSI('T', 'P', P, 'Q', 0, 'Water') for P in Prange]
Ssat1 = [PropsSI('S', 'P', P, 'Q', 0, 'Water') for P in Prange]
Tsat = [PropsSI('T', 'P', P, 'Q', 1, 'Water') for P in Prange]
Ssat = [PropsSI('S', 'P', P, 'Q', 1, 'Water') for P in Prange]

plt.plot(Ssat1, Tsat1, 'k-', linewidth=1.5)
plt.plot(Ssat, Tsat, 'k-', linewidth=1.5)

###plot b/w temp and entropy
S_actual = df_actualstates['S'].tolist() + [df_actualstates['S'][0]]
T_actual= df_actualstates['T'].tolist() + [df_actualstates['T'][0]]
S_ideal = df_idealstates['S'].tolist() + [df_idealstates['S'][0]]
T_ideal = df_idealstates['T'].tolist() + [df_idealstates['T'][0]]


plt.plot(S_actual, T_actual, 'r-', label='Actual Cycle')
plt.plot(S_ideal, T_ideal, 'b-', label='Ideal Cycle')
plt.legend()

plt.xlabel('Entropy (J/kg.K)')
plt.ylabel('Temperature (K)')
plt.title('Temperature vs Entropy of actual and ideal rankine cycles')
plt.show()
```
